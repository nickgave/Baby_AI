```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import csv
import pandas as pd
import os
```

```python
birth_rate = []
with open('./data/birth_rate.csv', 'r') as file:
    rdr = csv.reader(file)
    for line in rdr:
        if line[2] == '합계출산율(명)': continue
        birth_rate.append(float(line[2]))
    birth_rate = np.array(birth_rate)

print(birth_rate)
```

```python
folder_path = './data/'

file_names = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
data_list = []

# 각 파일을 읽고 numpy.array로 변환한 후, 리스트에 저장
for file_name in file_names:
    if file_name == "birth_rate.csv": continue
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)  # CSV 파일 읽기
    data_array = df.values[:,1]  # pandas DataFrame을 numpy array로 변환
    data_list.append(data_array)

data_list = np.column_stack((data_list[0], data_list[1], data_list[2]))
print(data_list)
```

```python
# 모델 구현
model2 = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(3,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1)
])

model2.compile(optimizer='adam', loss='mean_squared_error')
model2.fit(data_list, birth_rate, epochs=300, batch_size=20)

#모델 평가
loss = model2.evaluate(data_list, birth_rate)
print(f"Model loss: {loss:.10}")
```

```python
# 각 파일을 읽고 numpy.array로 변환한 후, 리스트에 저장
df = pd.read_csv('./data/educational_cost.csv')  # CSV 파일 읽기
data_array = df.values[:]  # pandas DataFrame을 numpy array로 변환
years = np.array(data_array[:,0])
Private_education  = np.array(data_array[:,1])

df = pd.read_csv('./data/house_cost.csv')  
data_array = df.values[:]  
house_cost  = np.array(data_array[:,1])

df = pd.read_csv('./data/marry_num.csv') 
data_array = df.values[:]  
marry_num  = np.array(data_array[:,1])

dataframe = pd.DataFrame({
    'Year': years,
    'education expenses': Private_education,
    'House price': house_cost,
    'Marrage Nums': marry_num
}); print(dataframe)
```

```python
#x축, y축 설정
x = dataframe['Year'].values
y = dataframe['education expenses'].values

#y=mx+n에서 최소제곱법을 이용해 m값, n값 구하기
m1 = np.cov(x,y,ddof=1)[0,1]/np.cov(x,y,ddof=1)[0,0]
n1 = np.mean(y)-m1*np.mean(x)

print(f"Regression Equation: {m1:.5f}x{n1:.5f}")

Edu = []

m1 = float(input())
n1 = -m1 * x[-1] + y[-1]
print(f"Changed Equation: {m1:.5f}x+{n1:.5f}")
a = 2024
while a <= 2040:
    Edu.append(m1 * a + n1)
    print(f'{a}년: {(Edu[-1]):.3}만 원')
    a += 1
```

```python
#x축, y축 설정
x = dataframe['Year'].values
y = dataframe['House price'].values

#y=mx+n에서 최소제곱법을 이용해 m값, n값 구하기
m2 = np.cov(x,y,ddof=1)[0,1]/np.cov(x,y,ddof=1)[0,0]
n2 = np.mean(y)-m2*np.mean(x)

print(f"Regression Equation: {m2:.5f}x{n2:.5f}")

House = []

m2 = float(input())
n2 = -m2 * x[-1] + y[-1]
print(f"Changed Equation: {m2:.5f}x+{n2:.5f}")
a = 2024
while a <= 2040:
    House.append(m2 * a + n2)
    print(f'{a}년: {(House[-1]):.5}')
    a += 1
```

```python
#x축, y축 설정
x = dataframe['Year'].values
y = dataframe['Marrage Nums'].values

#y=mx+n에서 최소제곱법을 이용해 m값, n값 구하기
m3 = np.cov(x,y,ddof=1)[0,1]/np.cov(x,y,ddof=1)[0,0]
n3 = np.mean(y)-m3*np.mean(x)

print(f"Regression Equation: {m3:.5f}x+{n3:.5f}")

Marrage = []

m3 = float(input())
n3 = -m3 * x[-1] + y[-1]
print(f"Changed Equation: {m3:.5f}x+{n3:.5f}")
a = 2024
while a <= 2040:
    Marrage.append(m3 * a + n3)
    print(f'{a}년: {(Marrage[-1]):.8}건')
    a += 1
```

```python
new_data = np.column_stack((Edu, House, Marrage))
print(new_data)
```

```python
# [사교육비(만 원), 지수(2017 = 100), 결혼(건)]
year = 2024
predictions = model2.predict(new_data)
print(predictions)
for prediction in predictions:
    print(f"{year} Predicted Birth Rate: {prediction[0]:.7}")
    year += 1
```
