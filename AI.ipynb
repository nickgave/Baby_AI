```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import csv
import pandas as pd
import os
```

```python
birth_rate = []
with open('./data/birth_rate.csv', 'r') as file:
    rdr = csv.reader(file)
    for line in rdr:
        if line[2] == '합계출산율(명)': continue
        birth_rate.append(float(line[2]))
    birth_rate = np.array(birth_rate)

print(birth_rate)
```

```python
folder_path = './data/'

file_names = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
data_list = []

# 각 파일을 읽고 numpy.array로 변환한 후, 리스트에 저장
for file_name in file_names:
    if file_name == "birth_rate.csv": continue
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)  # CSV 파일 읽기
    data_array = df.values[:,1]  # pandas DataFrame을 numpy array로 변환
    data_list.append(data_array)

data_list = np.column_stack((data_list[0], data_list[1], data_list[2]))
print(data_list)
```

```python
# 모델 구현
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(3,)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(data_list, birth_rate, epochs=150, batch_size=10)

#모델 평가
loss = model.evaluate(data_list, birth_rate)
print(f"Model loss: {loss:.10}")
```

```python
new_data = np.array([[30, 120, 400000]]) # [사교육비(만 원), 지수(2017 = 100), 결혼(건)]
predictions = model.predict(new_data)
for prediction in predictions:
    print(f"Predicted Birth Rate: {prediction[0]:.2}")
```
